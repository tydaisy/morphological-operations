```python
from pathlib import Path
from src.image.image import Image
from src.utils import imshow
from skimage import morphology
```

```python
image_original = Image(Path("../../data/demo1.png")).binary
imshow(image_original)
```

### 1. dilation with structuring element $3 \times 3$

```python
kernal = morphology.footprint_rectangle((3, 3))
image_dilated = morphology.dilation(image_original, kernal)
imshow(image_dilated)
```

### 2. dilation with structuring element $5 \times 5$

```python
kernal = morphology.footprint_rectangle((5, 5))
image_dilated = morphology.dilation(image_original, kernal)
imshow(image_dilated)
```

### 2. dilation with structuring element $7 \times 7$

```python
kernal = morphology.footprint_rectangle((7, 7))
image_dilated = morphology.dilation(image_original, kernal)
imshow(image_dilated)
```
