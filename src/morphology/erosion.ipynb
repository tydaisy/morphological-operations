```python
from pathlib import Path
from src.image.image import Image
import numpy as np
from matplotlib import pyplot as plt
from skimage import morphology
```

```python
def imshow(image: np.ndarray):
    """Show the image with matplotlib."""
    fig = plt.figure(frameon=False)

    # Add an axes that fills the whole figure
    ax = plt.Axes(fig, [0, 0, 1, 1])
    ax.set_axis_off()
    fig.add_axes(ax)

    ax.imshow(image, cmap='gray')
    plt.show()
```

```python
image_original = Image(Path("../../data/demo1.png")).binary
imshow(image_original)
```

### 1. erosion with structuring element $3 \times 3$

```python
kernal = morphology.footprint_rectangle((3, 3))
image_eroded = morphology.erosion(image_original, kernal)
imshow(image_eroded)
```

### 2. erosion with structuring element $5 \times 5$

```python
kernal = morphology.footprint_rectangle((5, 5))
image_eroded = morphology.erosion(image_original, kernal)
imshow(image_eroded)
```

### 2. erosion with structuring element $7 \times 7$

```python
kernal = morphology.footprint_rectangle((7, 7))
image_eroded = morphology.erosion(image_original, kernal)
imshow(image_eroded)
```
